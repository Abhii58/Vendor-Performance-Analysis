```python
import pandas as pd 
import os 
from sqlalchemy import create_engine, text
from urllib.parse import quote
import logging 
import time 

logging.basicConfig(
    filename="logs/ingestion_db.log", level = logging.DEBUG, format="%(asctime)s - %(levelname)s - %(message)s", filemode="a"
)

#engine = create_engine('mysql+pymysql://root:"ab@sql123"@127.0.0.1:3306/inventory')
password = quote('ab@sql123')

# Connect to mysql system DB first to create inventory DB if not exists
sys_engine = create_engine(f'mysql+pymysql://root:{password}@127.0.0.1/mysql')
with sys_engine.connect() as sys_conn:
    sys_conn.execute(text("CREATE DATABASE IF NOT EXISTS `inventory` CHARACTER SET utf8mb4 COLLATE utf8mb4_unicode_ci"))
    sys_conn.commit()
sys_engine.dispose()

# Now connect to inventory
engine = create_engine(f'mysql+pymysql://root:{password}@127.0.0.1:3306/inventory')
conn = engine.connect()
print("Connected successfully to inventory database!")

def ingest_db(df, table_name, engine):
    '''this func will ingest df into database table'''
    df.to_sql(table_name, con = engine, if_exists = 'replace', index = False)

#load the csv as dataframe  and ingest into db
def load_raw_data():
    start = time.time()
    for file in os.listdir('data'):
        if '.csv' in file:
             df = pd.read_csv('data/'+file)
             logging.info(f'ingesting (file) in db')
             print(df.shape)
             ingest_db(df, file[:-4], engine)
    end = time.time()
    total_time = (end - start)/60
    logging.info('---ingestion complete--------')
    logging.info('total time taken: (total_time)) minutes')

if __name__ == "__main__":
     load_raw_data()
```
